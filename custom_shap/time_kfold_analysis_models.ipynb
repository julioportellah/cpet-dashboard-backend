## The functions

```python
def interpret_model(model, data_df, features,label):
    modelLimitation = model.fit(data_df[features],data_df[label])
    explainer = shap.TreeExplainer(modelLimitation, data=data_df[features])
    shap_values = explainer.shap_values(data_df[features])
    shap.summary_plot(shap_values[1], data_df[features], plot_type='dot', max_display=data_df.shape[1])
    pass
```

```python
def get_waterfall_plot(model, data_df, features,label):
    modelLimitation = model.fit(data_df[features],data_df[label])
    explainer = shap.TreeExplainer(modelLimitation, data=data_df[features])
    shap_values = explainer.waterfall_plot(data_df[features])
    shap.summary_plot(shap_values[1], data_df[features], plot_type='dot', max_display=data_df.shape[1])
    pass
```

```python
def explain_model(model, df_data, features, ravel=False):
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(df_data[features])
    if ravel is True:
        for feature in features:
            shap.dependence_plot(feature, shap_values[1], df_data[features], show=False, interaction_index=None)
    else:
        fig, axes = plt.subplots(nrows=int(np.ceil(len(features)/4)), ncols=4, figsize=(20, 14))
        axes = axes.ravel()
        for i, col in enumerate(features):
            shap.dependence_plot(col, shap_values[1], df_data[features], ax=axes[i], show=False, interaction_index=None)
```

```python
def save_model(model, scaler, name):
    filename = name + '.sav'
    pickle.dump(model, open(filename,'wb'))
    filename = 'scaler_' + name + '.sav'
    pickle.dump(scaler, open(filename,'wb'))
```

```python
def balanced_k_fold_binary(X, y, nfold=5):
    fold_indexes = []
    indices = np.arange(X.shape[0])
    all_ones = y*(np.arange(X.shape[0])+1)
    all_ones = all_ones[all_ones!=0] - 1
    np.random.shuffle(all_ones)
    all_zeros = (1-y)*(np.arange(X.shape[0])+1)
    all_zeros = all_zeros[all_zeros!=0] - 1
    np.random.shuffle(all_zeros)
    one_size_sample = int(np.floor(all_ones.shape[0]/nfold))
    zero_size_sample = int(np.floor(all_zeros.shape[0]/nfold))
    kfold_list = []
    for i in range(nfold):
        if(i==(nfold-1)):
            mixed_value = np.array([])
            mixed_value = np.append(all_ones, all_zeros)
            np.random.shuffle(mixed_value)
            kfold_list.append(mixed_value.tolist())
        else:
            k_one_fold = all_ones[:one_size_sample]
            all_ones = all_ones[one_size_sample:]
            k_zero_fold = all_zeros[:zero_size_sample]
            all_zeros = all_zeros[zero_size_sample:]
            mixed_value = np.array([])
            mixed_value = np.append(k_one_fold, k_zero_fold)
            np.random.shuffle(mixed_value)
            #print(k_one_fold)

            kfold_list.append(mixed_value.tolist())
            pass
        pass
    return kfold_list
    pass
```

```python
def evaluate_model_using_balanced_kfold(X, y, clf_model, clf_scaler, nfolds=5):
    balanced_folds = balanced_k_fold_binary(X, y, nfolds)
    list_auc = []
    list_accuracy = []
    list_scores = []
    list_specificity = []
    list_sensitivity = []
    list_ppv = []
    for i in range(len(balanced_folds)):
        balanced_folds_copy = balanced_folds.copy()
        X_test = X[balanced_folds_copy[i]]
        y_test = y[balanced_folds_copy[i]]
        balanced_folds_copy.pop(i)
        train_indexes = list(np.concatenate(balanced_folds_copy).flat)
        X_train = X[train_indexes]
        y_train = y[train_indexes]
        sm = SMOTE()
        X_train_oversampled, y_train_oversampled = sm.fit_resample(X_train, y_train)
        clf_scaler.fit(X_train_oversampled)
        X_train_scaled = clf_scaler.transform(X_train_oversampled)
        X_test_scaled = clf_scaler.transform(X_test)
        clf_model.fit(X_train_scaled, y_train_oversampled)  
        y_pred = clf_model.predict_proba(X_test_scaled)
        y_pred = y_pred[:,1]
        fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
        tn, fp, fn, tp = confusion_matrix(y_test,np.round(y_pred)).ravel()
        auc_score = roc_auc_score(y_test, y_pred)
        accuracy_result = accuracy_score(y_test, np.round(y_pred))#clf_model.score(X_test, y_test)
        f1_score_result = f1_score(y_test, np.round(y_pred))
        specificity = tn / (tn+fp)
        sensitivity = tp / (tp+fn)
        ppv = tp/(tp+fp)
        print(f'For fold {i+1}:')
        print(f'AUC: {np.round(metrics.auc(fpr, tpr),3)}, Accuracy: {np.round(accuracy_result,3)}, f-score: {np.round(f1_score_result,3)}, Specificity: {np.round(specificity,3)}, Sensitivity: {np.round(sensitivity,3)}, PPV: {np.round(ppv,3)}')
        list_auc.append(auc_score)
        list_accuracy.append(accuracy_result)
        list_scores.append(f1_score_result)
        list_specificity.append(specificity)
        list_sensitivity.append(sensitivity)
        list_ppv.append(ppv)
        pass
    list_auc = np.array(list_auc)
    list_accuracy = np.array(list_accuracy)
    list_scores = np.array(list_scores)
    list_specificity = np.array(list_specificity)
    list_sensitivity = np.array(list_sensitivity)
    list_ppv = np.array(list_ppv)
    print("Report")
    print("AUC: {:.3f}+-{:.3f}".format(np.mean(list_auc),np.std(list_auc)))
    print("Accuracy: {:.3f}+-{:.3f}".format(np.mean(list_accuracy),np.std(list_accuracy)))
    print("F1 score: {:.3f}+-{:.3f}".format(np.mean(list_scores),np.std(list_scores)))
    print("Specificity: {:.3f}+-{:.3f}".format(np.mean(list_specificity),np.std(list_specificity)))
    print("Sensitivity: {:.3f}+-{:.3f}".format(np.mean(list_sensitivity),np.std(list_sensitivity)))
    print("PPV: {:.3f}+-{:.3f}".format(np.mean(list_ppv),np.std(list_ppv)))
    pass
```

### Creating the final models and adding interpretation layers

#### General Method

```python
def generate_model(X, y, clf_scaler, clf_model):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
    sm = SMOTE()
    X_train_oversampled, y_train_oversampled = sm.fit_resample(X_train, y_train)
    clf_scaler.fit(X_train_oversampled)
    X_train_scaled = clf_scaler.transform(X_train_oversampled)
    X_test_scaled = clf_scaler.transform(X_test)
    clf_model.fit(X_train_scaled, y_train_oversampled)
    y_pred = clf_model.predict_proba(X_test_scaled)
    y_pred = y_pred[:,1]
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    print(f"Model generated, AUC {np.round(metrics.auc(fpr, tpr),3)}")
    return clf_model, clf_scaler, y_test, y_pred, X_train, y_train, X_train_scaled, y_train_oversampled
```
